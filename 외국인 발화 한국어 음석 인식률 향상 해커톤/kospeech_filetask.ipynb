# 파일 옮기기

txt파일 옮기기

```python
import os
from glob import glob
import shutil

%cd /content/gdrive/MyDrive/Project/sound/NIA_contest/icanhearyourvoice/UnzipTxt

lists = glob('*.txt')

num = 1
for k in range(6) :
  for q in range(5) :
    sk = 'OurSpeech0'+str(k+1)
    if len(str(num))==1 : 
      sn = 'OurSpeech00'+str(num)
    else : 
      sn = 'OurSpeech0'+str(num)
    dir =  '/content/gdrive/MyDrive/Project/sound/NIA_contest/icanhearyourvoice/OurSpeech/'+sk+'/'+sn
    for j in range(0, 1000) :
        shutil.move(lists[0],dir)
        lists.remove(lists[0])
    num+=1
```

.wav를 .pcm으로 바꾸기

```python
import os
from glob import glob
import shutil

num = 1
for k in range(6) :
  for q in range(5) :
    sk = 'OurSpeech0'+str(k+1)
    if len(str(num))==1 : 
      sn = 'OurSpeech00'+str(num)
    else : 
      sn = 'OurSpeech0'+str(num)

    %cd /content/gdrive/MyDrive/Project/sound/NIA_contest/icanhearyourvoice/OurSpeech/{sk}/{sn}
    lists = glob('*.wav')
    for j in lists :
      li = j.split('.')
      new_filename = li[0]+'.pcm'
      os.rename(j, new_filename)
    num+=1
```

txt 파일명 wav와 동일하게 변경

```python
import os
from glob import glob
import shutil

num = 1
for k in range(6) :
  for q in range(5) :
    sk = 'OurSpeech0'+str(k+1)
    if len(str(num))==1 : 
      sn = 'OurSpeech00'+str(num)
    else : 
      sn = 'OurSpeech0'+str(num)

    %cd /content/gdrive/MyDrive/Project/sound/NIA_contest/icanhearyourvoice/OurSpeech/{sk}/{sn}
    lists = glob('*.txt')
    for j in lists :
      li = j.split('_')
      new_filename = li[0]+'_'+li[1]+'_'+li[2]+'.txt'
      os.rename(j, new_filename)
    num+=1
```
