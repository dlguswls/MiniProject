# 2. Json -> Text

```python
import os

PATH = 'E:/음성인식/data/train/'

JSON_PATH = os.path.join(PATH,'[train] 메타데이터_json')

TEXT_PATH = os.path.join(PATH,'[train] Data_txt')
```

```python
os.chdir(JSON_PATH)
```

```python
import json
from glob import glob


json_list = glob('*.json')

json_text = list()
```

```python
# json 파일의 txt부분만 가져오기

for i in range(len(json_list)):
    with open(f'{json_list[i]}',encoding='utf-8') as json_file:
        json_data = json.load(json_file)
        json_text.append(json_data['transcription']['ReadingLabelText'])
```

```python
for j in range(len(json_list)):  
  f = open(TEXT_PATH+f'/{json_list[j][:-5]}'+'.txt','w')
  data = json_text[j]
  f.write(data)
f.close()
```

```python
f = open(TEXT_PATH+f'/{json_list[299999][:-5]}'+'.txt','r')

line = f.readline()
print(line)
```
