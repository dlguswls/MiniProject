```python
# 양호(정상) 0
# 불량 1

for u in range(len(list_name)):
  try:
    %cd ./road/
    list_name = sorted(os.listdir())
    os.chdir(f'{list_name[u]}')
    os.chdir('1')
      
    json_list = []
    image_list = []

    for i in glob('*') : 
      if i in glob('*.jpg') : 
        image_list.append(i)
      elif i in glob('*.jpeg') : 
        image_list.append(i)
      elif i in glob('*.json') : 
        json_list.append(i)

    while len(json_list)!=len(image_list) : 
      for i in range(min(len(json_list), len(image_list))) : 
        if image_list[0][-4:]=='.jpg' : 
          if json_list[i][:-5]!=image_list[i][:-4] : 
            if image_list[i][:-4] in [k[:-5] for k in json_list]: 
              os.remove(patH + '/road/'+list_name[u]+'/1/'+json_list[i])
              break
            else :  
              os.remove(patH + '/road/'+list_name[u]+'/1/'+image_list[i])
              break
        elif image_list[0][-5:]=='.jpeg' :
          if json_list[i][:-5]!=image_list[i][:-5] : 
            if image_list[i][:-5] in [k[:-5] for k in json_list]: 
              os.remove(patH + '/road/'+list_name[u]+'/1/'+json_list[i])
              break
            else :  
              os.remove(patH + '/road/'+list_name[u]+'/1/'+image_list[i])
              break

      json_list = []
      image_list = []
      for i in glob('*') : 
        if i in glob('*.jpg') : 
          image_list.append(i)
        elif i in glob('*.jpeg') : 
          image_list.append(i)
        elif i in glob('*.json') : 
          json_list.append(i)

    data = []
    for i in range(len(json_list)):
      with open(json_list[i], encoding='UTF8') as data_file:
        data.append(json.load(data_file))

    check_list_1 = []
    for i in range(len(data)):
      check_list_1.append(data[i]['annotations'][0]['is_defect'])

    for i in range(len(check_list_1)):
      if check_list_1[i] =='정상':
        print(image_list[i], json_list[i])
        src = patH + '/road/'+list_name[u]+'/1'
        dir = patH + '/road/'+list_name[u]+'/0'

        shutil.move(str(json_list[i]), dir)
        shutil.move(str(image_list[i]), dir)
  except FileNotFoundError:
      pass
```

### 리스트 여러 개 이미지 자르기

```python
def img_cut_xy_no(filename, ny = '1') : 
  no_path = patH + '/road/'+filename+'/'+ny+'/'
  no_file_list = os.listdir(no_path)
  no_json_file = [file for file in no_file_list if file.endswith('.json')]

  import json
  no_dict_list = []

  for i in no_json_file : 
    for line in open((no_path+i),"r", encoding = 'utf-8') : 
      no_dict_list.append(json.loads(line))

  no_images_filename = [no_dict_list[i]['info']['filename'] for i in range(len(no_dict_list))]
  %cd {no_path}
  no_img = [cv2.imread(no_images_filename[i]) for i in range(len(no_images_filename))]
  no_img_expands = [no_dict_list[i]['annotations'] for i in range(len(no_dict_list))]
    
  for q in range(len(no_img_expands)) : 
    try : 
      for k in range(len(no_img_expands[q])) :
        no_img_expand = no_img_expands[q][k]['annotation_info'] # 순서대로 annotation 부분만 가져오기
        no_img_defected = no_img_expands[q][k]['is_defect'] # 순서대로 is_defected 부분만 가져오기

        x = []
        y = []

        for j in range(len(no_img_expand)) : 
          x.append(float(no_img_expand[j][0])) #x부분은 x에 추가
          y.append(float(no_img_expand[j][1])) #y부분은 y에 추가
        max_x = int(max(x))
        min_x = int(min(x))
        max_y = int(max(y))
        min_y = int(min(y))
        save_img = no_img[q][min_y : max_y, min_x : max_x].copy() #max값으로 자름
        x = []
        y = []

        if no_img_defected=='불량전체' : 
          %cd {patH}/road/{filename}/new/{ny}
          name = no_images_filename[q].split('.')[0]+'__all_'+str(k)+'.jpg'
          cv2.imwrite(name,save_img)
        elif no_img_defected=='불량부분' : 
          %cd {patH}/road/{filename}/new/part
          name = no_images_filename[q].split('.')[0]+'__part_'+str(k)+'.jpg'
          cv2.imwrite(name,save_img)

        %cd {patH}/road/{filename}/{ny}

    except KeyError: 
      print(no_images_filename[q])
    except TypeError : 
      print(no_images_filename[q])
```

```python
def img_cut_xy_yes(filename, ny = '0') : 
  no_path = patH + '/road/'+filename+'/'+ny+'/'
  no_file_list = os.listdir(no_path)
  no_json_file = [file for file in no_file_list if file.endswith('.json')]

  import json
  no_dict_list = []

  for i in no_json_file : 
    for line in open((no_path+i),"r", encoding = 'utf-8') : 
      no_dict_list.append(json.loads(line))

  no_images_filename = [no_dict_list[i]['info']['filename'] for i in range(len(no_dict_list))]
  %cd {no_path}
  no_img = [cv2.imread(no_images_filename[i]) for i in range(len(no_images_filename))]
  no_img_expands = [no_dict_list[i]['annotations'] for i in range(len(no_dict_list))]
  
  for q in range(len(no_img_expands)) : 
    try : 
      for k in range(len(no_img_expands[q])) :
        no_img_expand = no_img_expands[q][k]['annotation_info']

        x = []
        y = []

        for j in range(len(no_img_expand)) : 
          x.append(float(no_img_expand[j][0]))
          y.append(float(no_img_expand[j][1]))
        max_x = int(max(x))
        min_x = int(min(x))
        max_y = int(max(y))
        min_y = int(min(y))
        save_img = no_img[q][min_y : max_y, min_x : max_x].copy()
        x = []
        y = []

        %cd {patH}/road/{filename}/new/{ny}
        name = no_images_filename[q].split('.')[0]+'__'+str(k)+'.jpg'
        cv2.imwrite(name,save_img)
        %cd {patH}/road/{filename}/{ny}

    except KeyError: 
      print(no_images_filename[q])
    except TypeError : 
      print(no_images_filename[q])
```
